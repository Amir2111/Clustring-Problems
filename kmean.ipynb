```
import pandas as pd
dataset = pd.read_csv("/content/sample_data/Iris(1).csv")
x_input = dataset.iloc[:, 0:3]
print(x_input)
from sklearn.cluster import KMeans
clf = KMeans(n_clusters=3)
clf.fit(x_input)
centeroid = clf.cluster_centers_
pred = clf.predict(x_input)
print(pred)
print(centeroid)
```
